```python
# 學生總分為300
#有些學生可以加分5%

while True:
    try:
        scores = int(input("請輸入學生分數(最高300分):"))
        break
    except ValueError:
        print("輸入錯誤，請輸入整數！")

is_add = input("學生是否符合加分條件?(y,n)")

if is_add == "y": #單項選擇
	scores = scores * 1.05

# 四捨五入(5進位)
scores = int(scores + 0.5)

# 超過300分，上限取300
if scores > 300:
	scores = 300

print("學生分數:",scores)        
```
